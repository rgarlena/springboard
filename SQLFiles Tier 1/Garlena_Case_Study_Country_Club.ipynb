```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn
```

/* Q1: Some of the facilities charge a fee to members, but some do not.
Write a SQL query to produce a list of the names of the facilities that do. */

```python
def facilities_fee(conn):
    """
    Query facilities table for facilities names that charge fee to members 
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT name 
        FROM Facilities 
        WHERE membercost > 0;
        """
    cur.execute(query1)
 
    col_names = [description[0] for description in cur.description]
    print("Column Name:", col_names)
    
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("1. Query facilties that charge member fee")
        facilities_fee(conn)
 
 
if __name__ == '__main__':
    main()
```

/* Q3: Write an SQL query to show a list of facilities that charge a fee to members,
where the fee is less than 20% of the facility's monthly maintenance cost.
Return the facid, facility name, member cost, and monthly maintenance of the
facilities in question. */

```python
def monthy_maintenance(conn):
    """
    Query facilities where member fee is less than 20% of the facility monthly maintenance cost. 
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query3 = """
        SELECT facid, 
           name AS facility_name, 
           membercost, 
           monthlymaintenance 
        FROM Facilities 
        WHERE membercost < 0.2 * monthlymaintenance;
        """
    cur.execute(query3)
    
    col_names = [description[0] for description in cur.description]
    print("Column Names:", col_names)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("3. Query facilities where member fee is less than 20% of the facility monthly maintenance cost")
        monthy_maintenance(conn)
 
if __name__ == '__main__':
    main()
```

/* Q5: Produce a list of facilities, with each labelled as
'cheap' or 'expensive', depending on if their monthly maintenance cost is
more than $100. Return the name and monthly maintenance of the facilities
in question. */

```python
def monthly_maintenance_cost(conn):
    """
    Query facilities and categorize them as 'expensive' or 'cheap' based on their monthly maintenance cost.
    :param conn: the Connection object
    :return: None
    """
    cur = conn.cursor()
    
    query = """
        SELECT name, monthlymaintenance,
            CASE
                WHEN monthlymaintenance > 100 THEN 'expensive'
                ELSE 'cheap'
            END AS maintenance_cost_category
        FROM Facilities;
        """
    cur.execute(query)
    
    col_names = [description[0] for description in cur.description]
    print("Column Names:", col_names)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)

def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("5. Query facilities with cost over and under $100")
        monthly_maintenance_cost(conn)
 
if __name__ == '__main__':
    main()
```

/* Q7: Produce a list of all members who have used a tennis court.
Include in your output the name of the court, and the name of the member
formatted as a single column. Ensure no duplicate data, and order by
the member name. */

```python
def members_tennis(conn):
    """
    Query members who have used a tennis court. 
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query7 = """
        SELECT DISTINCT 
            (m.firstname || ' ' || m.surname) AS member_name,
            f.name AS court_name
        FROM Bookings AS b
        JOIN 
            Members AS m ON b.memid = m.memid
        JOIN 
            Facilities AS f ON b.facid = f.facid
        WHERE 
            f.name LIKE 'Tennis Court%'
        ORDER BY 
            member_name;
        """
    cur.execute(query7)
    
    col_names = [description[0] for description in cur.description]
    print("Column Names:", col_names)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("7. Query members who have used tennis court")
        members_tennis(conn)
 
if __name__ == '__main__':
    main()
```

/* Q8: Produce a list of bookings on the day of 2012-09-14 which
will cost the member (or guest) more than $30. Remember that guests have
different costs to members (the listed costs are per half-hour 'slot'), and
the guest user's ID is always 0. Include in your output the name of the
facility, the name of the member formatted as a single column, and the cost.
Order by descending cost, and do not use any subqueries. */

```python
def bookings_2012_09_14(conn):
    """
    Query bookings on 2012-09-14 costing over $30. 
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
     
    query8 = """
        SELECT 
            f.name AS facility_name,
            CASE
                WHEN m.memid = 0 THEN 'Guest'
                ELSE m.firstname || ' ' || m.surname
            END AS member_name,
            CASE
                WHEN m.memid = 0 THEN b.slots * f.guestcost
                ELSE b.slots * f.membercost
            END AS cost
        FROM 
            Bookings AS b
        JOIN 
            Facilities AS f ON b.facid = f.facid
        LEFT JOIN 
            Members AS m ON b.memid = m.memid
        WHERE 
            b.starttime LIKE '2012-09-14%'
            AND (
            (m.memid = 0 AND b.slots * f.guestcost > 30)
            OR (m.memid != 0 AND b.slots * f.membercost > 30)
            )
        ORDER BY 
            cost DESC;
        """
    cur.execute(query8)
    
    col_names = [description[0] for description in cur.description]
    print("Column Names:", col_names)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("8. Query members who have used tennis court")
        bookings_2012_09_14(conn)
 
if __name__ == '__main__':
    main()
```

/* Q9: This time, produce the same result as in Q8, but using a subquery. */

```python
def bookings_2012_09_14_subquery(conn):
    """
    Query bookings on 2012-09-14 costing over $30 using a subquery. 
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
     
    query9 = """
    SELECT 
        f.name AS facility_name,
        CASE
            WHEN m.memid = 0 THEN 'Guest'
            ELSE m.firstname || ' ' || m.surname
        END AS member_name,
        cost
    FROM 
        (
            SELECT 
                b.facid,
                b.memid,
                CASE
                    WHEN b.memid = 0 THEN b.slots * f.guestcost
                    ELSE b.slots * f.membercost
                END AS cost
            FROM 
                Bookings AS b
            JOIN 
                Facilities AS f ON b.facid = f.facid
            WHERE 
                b.starttime LIKE '2012-09-14%'
                AND (
                    (b.memid = 0 AND b.slots * f.guestcost > 30)
                    OR (b.memid != 0 AND b.slots * f.membercost > 30)
                )
        ) AS subquery
    JOIN 
        Facilities AS f ON subquery.facid = f.facid
    LEFT JOIN 
        Members AS m ON subquery.memid = m.memid
    ORDER BY 
        cost DESC;

        """
    cur.execute(query9)
    
    col_names = [description[0] for description in cur.description]
    print("Column Names:", col_names)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("9. Query members who have used tennis court")
        bookings_2012_09_14_subquery(conn)
 
if __name__ == '__main__':
    main()
```

/* Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members! */

```python
def total_revenue_under_1000(conn):
    """
    Query facilities with total revenue less than $1000. 
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
     
    query10 = """
    SELECT 
        f.name AS facility_name,
        SUM(
            CASE
                WHEN b.memid = 0 THEN b.slots * f.guestcost
                ELSE b.slots * f.membercost
            END
        ) AS total_revenue
    FROM 
        Bookings AS b
    JOIN 
        Facilities AS f ON b.facid = f.facid
    GROUP BY 
        f.name
    HAVING 
        total_revenue < 1000
    ORDER BY 
        total_revenue;

        """
    cur.execute(query10)
    
    col_names = [description[0] for description in cur.description]
    print("Column Names:", col_names)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("10. Query facilities with total revenue less than $1000")
        total_revenue_under_1000(conn)
 
if __name__ == '__main__':
    main()
```

/* Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order */

```python
def recommended_by(conn):
    """
    Query members and who recommended them. 
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
     
    query11 = """
    SELECT 
        m.firstname || ' ' || m.surname AS member_name,
        r.firstname || ' ' || r.surname AS recommender_name
    FROM 
        Members AS m
    LEFT JOIN 
        Members AS r ON m.recommendedby = r.memid
    ORDER BY 
        m.surname, m.firstname;

        """
    cur.execute(query11)
    
    col_names = [description[0] for description in cur.description]
    print("Column Names:", col_names)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("11. Query members and who recommended them")
        recommended_by(conn)
 
if __name__ == '__main__':
    main()
```

/* Q12: Find the facilities with their usage by member, but not guests */

```python
def recommended_by(conn):
    """
    Query facilities usage by member. 
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
     
    query12 = """
    SELECT f.name AS facility_name, 
       COUNT(b.slots) AS member_usage 
    FROM Facilities AS f 
    JOIN Bookings AS b ON f.facid = b.facid 
    WHERE b.memid != 0 
    GROUP BY f.name;
        """
    cur.execute(query12)
    
    col_names = [description[0] for description in cur.description]
    print("Column Names:", col_names)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("12. Query facilities usage by member")
        recommended_by(conn)
 
if __name__ == '__main__':
    main()
```
